```python
#This is a script to produce parameter estimate figures that appear in paper: Neural Likelihood Surfaces.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
#Load exact and neural likelihood estimators for the evaluation data for single realizations
local_folder = "/home/juliatest/Dropbox/likelihood_free_inference/neural_likelihood/gaussian_process"
image_size = 25
image_name = str(image_size) + "_by_" + str(image_size)
version = "final_version"
neural_likelihood_estimators_file = (local_folder + "/evaluate_nn/produce_neural_likelihood_estimates/data/25_by_25/" +
                                     version + "/single/reps/200/neural_likelihood_estimators_9_by_9_density_" +
                                     image_name + "_image_200.npy")
exact_likelihood_estimators_file = (local_folder + "/evaluate_nn/produce_exact_likelihood_estimates/data/" 
                                    + image_name + "/single/reps/200/evaluation_ll_estimators_9_by_9_" + 
                                    image_name + "_image_200.npy")
#first column is variance and second is lengthscale
parameter_matrix = np.load((local_folder + "/evaluate_nn/generate_data/data/" + image_name + 
                            "/single/reps/200/evaluation_parameters_9_by_9_density_" + image_name + "_200.npy"))
parameter_matrix = parameter_matrix[:,0,:]
neural_likelihood_estimators = np.load(neural_likelihood_estimators_file)
exact_likelihood_estimators = np.load(exact_likelihood_estimators_file)
number_of_reps = 200
```

```python
#Load exact and neural likelihood estimators for the evaluation data for multiple realizations
multi_neural_likelihood_estimators_file = (local_folder + "/evaluate_nn/produce_neural_likelihood_estimates/data/25_by_25/" +
                                     version + "/multi/5/reps/200/neural_likelihood_estimators_9_by_9_density_" +
                                     image_name + "_multi_5_200.npy")
multi_exact_likelihood_estimators_file = (local_folder + "/evaluate_nn/produce_exact_likelihood_estimates/data/" 
                                    + image_name + "/multi/5/reps/200/evaluation_ll_estimators_9_by_9_image_" 
                                    + image_name + "_multi_5_200.npy")
#first column is variance and second is lengthscale
parameter_matrix = np.load((local_folder + "/evaluate_nn/generate_data/data/" + image_name + 
                            "/single/reps/200/evaluation_parameters_9_by_9_density_" + image_name + "_200.npy"))
parameter_matrix = parameter_matrix[:,0,:]
multi_neural_likelihood_estimators = np.load(multi_neural_likelihood_estimators_file)
multi_exact_likelihood_estimators = np.load(multi_exact_likelihood_estimators_file)
number_of_reps = 200
```

```python
#These are the indices in the parameter matrix which correspond to 4 by 4 grid over the parameter space
indices = [64,66,68,70,46,48,50,52,28,30,32,34,10,12,14,16]
parameter_matrix = parameter_matrix[indices,:]
#The parameter estimates for the 4 by 4 grid
exact_likelihood_estimators = exact_likelihood_estimators[indices,:,:]
neural_likelihood_estimators = neural_likelihood_estimators[indices,:,:]
multi_exact_likelihood_estimators = multi_exact_likelihood_estimators[indices,:,:]
multi_neural_likelihood_estimators = multi_neural_likelihood_estimators[indices,:,:]
jittered_exact_likelihood_estimators = exact_likelihood_estimators + np.random.normal(0, .02, (16, number_of_reps, 2))
jittered_neural_likelihood_estimators = neural_likelihood_estimators + np.random.normal(0, .02, (16, number_of_reps, 2))
jittered_multi_exact_likelihood_estimators = multi_exact_likelihood_estimators + np.random.normal(0, .02, (16, number_of_reps, 2))
jittered_multi_neural_likelihood_estimators = multi_neural_likelihood_estimators + np.random.normal(0, .02, (16, number_of_reps, 2))
```

```python
def visualize_exact_and_neural_likelihood_estimates(exact_likelihood_estimators, neural_likelihood_estimators,
                                                    parameter_matrix, realization_type, rounding_digits):
    fig, ax = plt.subplots(nrows = 4, ncols = 4, figsize = (15, 15), sharex = True, sharey = True, squeeze = True)
    plt.xlim(0,2)
    plt.ylim(0,2)
    for i in range(0,4):
        for j in range(0, 4):
            ax[i,j].scatter(jittered_neural_likelihood_estimators[(i*4+j),:,0],
                            jittered_neural_likelihood_estimators[(i*4+j),:,1], s = 1.5, marker = 'o', c = 'lime')
            ax[i,j].scatter(jittered_exact_likelihood_estimators[(i*4+j),:,0],
                            jittered_exact_likelihood_estimators[(i*4+j),:,1], s = 1.5, marker = 'o', c = 'purple')
            #first is variance?
            true_variance = parameter_matrix[(i*4+j),0]
            true_lengthscale = parameter_matrix[(i*4+j),1]
            params = np.repeat(np.matrix([[true_lengthscale, true_variance]]), repeats = number_of_reps, axis = 0)
            ax[i,j].plot(true_lengthscale, true_variance, marker = '*', markersize = 12, c = 'black')
            ll_mse = np.round((np.linalg.norm(exact_likelihood_estimators[(i*4+j),:,:]-params)**2)/params.shape[0],
                              rounding_digits)
            psi_mse = np.round((np.linalg.norm(neural_likelihood_estimators[(i*4+j),:,:]-params)**2)/params.shape[0],
                               rounding_digits)
            labels = ["exact {}".format(ll_mse), "neural {}".format(psi_mse)]
            if((i==0) and (j == 0)):
                labels = ["exact mse {}".format(ll_mse), "neural mse {}".format(psi_mse)]
                legend_elements = [Line2D([0], [0], marker='o', color='w',
                                markerfacecolor='purple', markersize=17), 
                                Line2D([0], [0], marker='o', color='w',
                                markerfacecolor='lime', markersize=17)]
                ax[i,j].legend(handles=legend_elements, labels = labels, fontsize = 16, loc='lower right')
            else:
                labels = ["{}".format(ll_mse), "{}".format(psi_mse)]
                legend_elements = [Line2D([0], [0], marker='o', color='w',
                                markerfacecolor='purple', markersize=12), 
                                Line2D([0], [0], marker='o', color='w',
                                markerfacecolor='lime', markersize = 12)]
                ax[i,j].legend(handles=legend_elements, labels = labels, fontsize = 15)

            if(j == 0):
                ax[i,j].set_yticks([i*.4 for i in range(1, 5)])
                ax[i,j].set_yticklabels([round(.4*i, 1) for i in range(1, 5)], size=20)

            if(i == 3):
                ax[i,j].set_xticks([i*.4 for i in range(1, 5)])
                ax[i,j].set_xticklabels([round(.4*i, 1) for i in range(1, 5)], size=20)
    plt.subplots_adjust(wspace=.1, hspace=.05, left = .1, right = .98, bottom = .1, top = .94)
    fig.text(0.55, 0.04, 'length scale', ha='center', va='center', fontsize = 40)
    fig.text(0.04, 0.52, 'variance', ha='center', va='center', rotation='vertical', fontsize = 40)
    fig.text(0.15, .97, "Neural vs Exact Likelihood Estimates", fontsize = 45)
    fig_name = (local_folder + 
            "/evaluate_nn/visualizations/visualize_parameter_estimates/visualizations/4_by_4/" 
            + realization_type + "/reps/200/exact_and_neural_likelihood_estimators_" + image_name + "_200.png")
    plt.savefig(fig_name)
```

```python
realization_type = "single"
rounding_digits = 2
visualize_exact_and_neural_likelihood_estimates(exact_likelihood_estimators, neural_likelihood_estimators, parameter_matrix, realization_type,
                                                rounding_digits)
```

```python
realization_type = "multi/5"
rounding_digits = 3
visualize_exact_and_neural_likelihood_estimates(multi_exact_likelihood_estimators, multi_neural_likelihood_estimators,
                                                parameter_matrix, realization_type, rounding_digits)
```
